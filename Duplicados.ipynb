```python

import pandas as pd

##Aqu saca el el path de tu archivo a comparar

event1 = "/content/subir este mero 1.csv" # Aqu pon el archivo que presenta duplicados
event2 = "/content/subir este tambien.csv" # Aqu pon el archivo del evento con el que lo quieres comparar

# Cargar los datos en DataFrames de pandas
df1 = pd.read_csv(event1)
df2 = pd.read_csv(event2)

# Crear una nueva columna en df1 para marcar duplicados
df1['Duplicado'] = df1.apply(lambda row: 'duplicado' if any(df2[['ZONA', 'SECCIN', 'ASIENTO']].eq(row[['ZONA', 'SECCIN', 'ASIENTO']]).all(axis=1)) else '', axis=1)

# Show duplicate
duplicate = df1[df1['Duplicado'] == 'duplicado']
print("Duplicados encontrados:")
print(duplicate)

# Guardar los cambios en el archivo CSV del primer evento
df1.to_csv(event1, index=False)


```
